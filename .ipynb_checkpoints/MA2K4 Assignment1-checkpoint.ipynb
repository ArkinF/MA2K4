### 1.7

Let's define the $x_{+}$ and $x_{-}$ functions

```python
def x_plus(p_val, q_val):
    '''insert computation'''
    result = -p_val + np.sqrt(p_val**2 + q_val)
    return result
    
def x_minus(p_val, q_val):
    '''insert computation'''
    result = -p_val - np.sqrt(p_val**2 + q_val)
    return result
```

#### a)

Similarly for $R_{+}$ and $R_{-}$

```python
def R_plus_p(p_val, delta_p, q_val):
    result = ((x_plus(p_val+delta_p,q_val)-x_plus(p_val,q_val))/x_plus(p_val,q_val)) / (delta_p/p_val)
    return result

def R_minus_p(p_val, delta_p, q_val):
    result = ((x_minus(p_val+delta_p,q_val)-x_minus(p_val,q_val))/x_minus(p_val,q_val)) / (delta_p/p_val)
    return result
```

Now let's calculate the $R_{+}$ and $R_{-}$ values for $q=1$

```python
del_p = 1e-3
q = 1

for p in [-1, 1]:
    print(f'R_plus for p={p}, q={q}: {R_plus_p(p, del_p, q)}')
    print(f'R_minus for p={p}, q={q}: {R_minus_p(p, del_p, q)}')
```

Since these changes are all around 0.7, this indicates that the solution changes slightly less than the the perturbation. Hence the equation is not particularly sensitive to perturbations.

#### b)

First let's recall the analytic condition numbers $K(p)$ obtained from 1.4)a):

```python
def k_p(p,q):
    result = abs(p/np.sqrt(p**2+q))
    return result
```

Now let's plot the values of $R_+$ and $R_-$ for different values of $pâˆˆ[-1,1]$ with $K(p)$.

```python
# define variables
del_p = 1e-3
q = 1
p_vals = np.linspace(-1,1,200)

# plot R plus
plt.plot(p_vals, R_plus_p(p_vals, del_p, q), color='b', linewidth=2.5, label='$R_+$')
plt.plot(p_vals, k_p(p_vals,q), color='black', label='$K(p)$')
plt.ylabel('R+')
plt.xlabel('p')
plt.legend()
plt.show()

# plot R minus
plt.plot(p_vals, R_minus_p(p_vals, del_p, q), color='r', linewidth=2.5, label='$R_-$')
plt.plot(p_vals, k_p(p_vals,q), color='black', label='$K(p)$')
plt.ylabel('R-')
plt.xlabel('p')
plt.legend()
plt.show()
```

Here we can see that both $R$ functions are relatively linear, flattening near the endpoints -1 and 1. We can also see that $R_{+}$ and $R_{-}$ are mirror images of each other (in the y-axis). Additionally, we see that $K(p)$ is equal to $R_{+}$ for $p<0$ and equal to $R_{-}$ for $p>0$.

#### c)

Now let's repeat the process for $q$. First the $R_{+}$ and $R_{-}$ functions

```python
def R_plus_q(q_val, delta_q, p_val):
    return ((x_plus(p_val,q_val+delta_q)-x_plus(p_val,q_val))/x_plus(p_val,q_val)) / (delta_q/q_val)

def R_minus_q(q_val, delta_q, p_val):
    return ((x_minus(p_val,q_val+delta_q)-x_minus(p_val,q_val))/x_minus(p_val,q_val)) / (delta_q/q_val)
```

Then the $R_{+}$ and $R_{-}$ values, this time for $p=1$

```python
del_q = 1e-3
p = 1

for q in [-1, 1]:
    print(f'R_plus for p={p}, q={q}: {R_plus_q(q, del_q, p)}')
    print(f'R_minus for p={p}, q={q}: {R_minus_q(q, del_q, p)}')
    
```

Notice how when $q=-1$, the values are very large. This indicates that the equation is very sensitive for $q=-1$. When $q=1$, the equation is once again not particularly sensitive.

#### d)

Recall $K(q)$ from Q1.4)a). This time it varies for the positive and negative solutions.

```python
def k_q_plus(p,q):
    result = 0.5*abs(q/np.sqrt(p**2+q))*abs(1/(-p+np.sqrt(p**2+q)))
    return result

def k_q_minus(p,q):
    result = 0.5*abs(q/np.sqrt(p**2+q))*abs(1/(-p-np.sqrt(p**2+q)))
    return result
```

Now let's plot the values of $R_+$ and $R_-$ for different values of $qâˆˆ[-1,1]$ with $K(q)$.

```python
# define variables
del_q = 1e-3
p = 1
q_vals = np.linspace(-1,1,200)

# plot R plus
plt.plot(q_vals, R_plus_q(q_vals, del_q, p), color='b', linewidth=2.5, label='$R_+$')
plt.plot(q_vals, k_q_plus(p,q_vals), color='black', label='$K(q)$')
plt.ylabel('R+')
plt.xlabel('q')
plt.legend()
plt.show()

# plot R minus
plt.plot(q_vals, R_minus_q(q_vals, del_q, p), color='b', linewidth=2.5, label='$R_+$')
plt.plot(q_vals, k_q_minus(p,q_vals), color='black', label='$K(q)$')
plt.ylabel('R-')
plt.xlabel('q')
plt.legend()
plt.show()
```

In this case, we see that $R_{+}$ and $R_{+}$ are mirror images of each other in the x-axis. For positive $q$, we can see that both the $R$ terms are relatively flat, however they begin to diverge as $q$ tends to -1. The behaviour of $K(q)$ is also similar, going to infinity as $q\rightarrow-1$ and undefined at $q=-1$. This is because of the denominator of $K(q)$ tending to 0 here.

### 1.8

#### a)

Let's define the linear_interpolate function

```python
def linear_interpolate(f, x_0, x_1, x):
    gradient = (f(x_1)-f(x_0))/(x_1-x_0)
    result = f(x_0) + gradient*(x-x_0)
    return result
```

Let's evaluate the linear interpolant of $2sin2x$ with nodes at 0 and 1 at 0.75:

```python
def f_x(x):
    return 2*np.sin(2*x)

print(f'linear interpolant at 0.75: {linear_interpolate(f_x, 0, 1, 0.75)}')
```

#### b)

Let's plot $sinx$ in this interval along with its interpolant for reference

```python
x_vals = np.linspace(0,1,1000)

plt.plot(x_vals, np.sin(x_vals), label='sinx')
plt.plot(x_vals, linear_interpolate(np.sin, 0, 1, x_vals), label='interpolant')
plt.legend()
plt.show()
```

We can see that for this function, the linear interpolant is a pretty good approximation.

#### c)

Now let's work out the interpolation error for $logx$

```python
def interpolation_error(f, x_0, x_1, x):
    result = f(x) - linear_interpolate(f, x_0, x_1, x)
    return result
```

```python
x_0 = 10
x_1 = 11

x_vals = np.linspace(x_0,x_1,1000)

def f_x(x):
    return np.log(x)

plt.plot(x_vals, interpolation_error(f_x, x_0, x_1, x_vals))
plt.title('Interpolation error for $logx$')
plt.xlabel('x')
plt.ylabel('error')
plt.show()

# plt.plot(x_vals, f_x(x_vals))
# plt.plot(x_vals, linear_interpolate(f_x,x_0,x_1,x_vals))
# plt.show()

print(f'error at x=10.2: {interpolation_error(f_x, x_0, x_1, 10.2)}')
print(f'error at x=10.8: {interpolation_error(f_x, x_0, x_1, 10.8)}')
```

#### d)

Let's compute the maximum absolute error for the linear interpolation

```python
def max_error(f, x_0 , x_1):
    
    x_vals = np.linspace(x_0,x_1,1000)
    result = [max(abs(interpolation_error(f, x_0, x_1, x)) for x in x_vals)]
    return result[0]
```

```python
def f_x(x):
    return 1/(1+x**2)

h=1

print(f'Max error for 1/(1+x^2): {max_error(f_x,0,h)}')
```

Now let's compute the max error for a range of $h$ values over $[0,1]$. We shall plot this on a log-log graph (base 10).

```python
h_vals = np.array([2**-k for k in range(11)])
h_vals

# I have used base 10 for the log-log plot
plt.plot(np.log10(h_vals), np.log10([max_error(f_x, 0, h) for h in h_vals]))
plt.title('max error for h')
plt.xlabel('log$_{10}$(h)')
plt.ylabel('log$_{10}$(max error)')
plt.show()
```

We observe that the $log(max error)$ is linear to $log(h)$, implying that the error is proportional to some polynomial of $h$. Let's work out the order of the proportional polynomial.

```python
plt.plot(np.log10(h_vals), np.log10([max_error(f_x, 0, h) for h in h_vals]), label='max error')
plt.plot(np.log10(h_vals), np.log10(h_vals**2), label='$h^2$')
plt.xlabel('log$_{10}$(h)')
plt.ylabel('log$_{10}$(y)')
plt.legend()
plt.show()
```

Adding the curve $h^{\gamma}$, we find that a $\gamma$ of 2 gives us parallel lines (matching decay rates). This implies that the error scales quadratically with $h$.

### 1.9

#### a)

Let's begin by defining a function to calculate the k-th Lagrange polynomial.

```python
def lagrange(k, x, nodes):
    
    L_x = 1
    for j in range(len(nodes)):
        if j != k:
            L_x *= (x-nodes[j])/(nodes[k]-nodes[j])

    return L_x
```

Now let's plot some Lagrange polynomials for $xâˆˆ[0,4]$

```python
nodes = [0,1,2,3,4]
x_vals = np.linspace(0,4,400)

for i in nodes:
    plt.plot(x_vals, lagrange(i, x_vals, nodes), label=f'k={i}')

plt.xlabel('x values')
plt.ylabel('function values')
plt.title('Lagrange Polynomials')
plt.legend(title='Nodes:')
plt.show()
```

Similarly for $xâˆˆ[0,10]$

```python
nodes = np.arange(11)
x_vals = np.linspace(0,10,1000)

for i in nodes:
    plt.plot(x_vals, lagrange(i, x_vals, nodes), label=f'k={i}')

plt.xlabel('x values')
plt.ylabel('function values')
plt.title('Lagrange Polynomials')
plt.legend(title='Nodes:')
plt.show()
```

#### b)

Now let's perform the Lagrange interpolation for $1/(1+x^2)$. We will use m equally spaced nodes over the interval $[-5,5]$.

```python
def lagrange_interp(x, nodes, values):
    
    p_x = 0
    for k in range(len(nodes)):
        p_x += values[k]*lagrange(k,x,nodes)
    return p_x

def f_x(x):
    return 1/(1+x**2)
```

```python
m_vals = [3,5,11,21]
x_vals = np.linspace(-5,5,1000)
f_vals = f_x(x_vals)

plt.figure(figsize=(10, 8))

for i, m in enumerate(m_vals):
    nodes = np.linspace(-5, 5, m)
    vals = f_x(nodes)

    plt.subplot(3, 2, i + 1)
    plt.plot(x_vals, lagrange_interp(x_vals, nodes, vals), color='red')
    plt.plot(x_vals, f_vals, color='blue')
    plt.plot(nodes, vals, '.', color='black')
    
    plt.title(f'Interpolation with m={m}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    
plt.tight_layout()
plt.show()
```

It seems that greater values of $m$ lead to greater accuracy for the main section of the polynomial as it passes through more points. However, this also results in greater divergence at the endpoints. This is very evident for the $m=21$ polynomial which is pretty accurate from $-4<x<4$ but blows out of proportion at the ends of the $x$-interval.

This is probably because these higher, more accurate polynomials have much greater coefficients of $x$, which causes extreme oscillations as $|x|$ increases. This is known as $\textbf{Runge's phenomenon}$.

Let's also compute the polynomials at $x=1$ and $x=4.5$ and compare to the actual values.

```python
x_1={}
x_2={}

for m in m_vals:
    nodes = np.linspace(-5,5,m)
    vals = f_x(nodes)
    x_1[m] = lagrange_interp(1, nodes, vals)
    x_2[m] = lagrange_interp(4.5, nodes, vals)

df_1 = pd.DataFrame({'1':x_1})
df_2 = pd.DataFrame({'4.5': x_2})

combined = pd.concat([df_1, df_2], axis=1)

combined.loc['actual'] = [f_x(1), f_x(4.5)]

display(combined)
```

For $x=1$, the values of the polynomial consistently get better as $m$ increases, until they are exactly correct for $m=11$ and $m=21$ (since 1 is a node here). However for $x=4.5$, we notice that the values actually get worse as $m$ increases until we have 4.5 as a node for $m=21$. This illustrates how, for big $m$, a Lagrange polynomial oscillates more and more near the endpoints of a region.

Weierstrass's theorem states that for $f$ continuous on $[a,b]$, there exists an arbitrarily close polynomial approximation such that $|f(x)-p(x)|<Îµ$ for all $xâˆˆ[a,b]$

These observations relate because we can see that as the number of nodes increase, the polynomial gets closer and closer to $f$. This is more obvious for a point near the centre like $x=1$ but for $x=4.5$, the values vary more before getting accurate. This indicates how important the choice of nodes is, especially near the endpoints of the interval.
