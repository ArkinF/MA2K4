```python
def scientific(a, precision=3):
    expo=int(np.log(abs(a))/np.log(10))
    expo=expo if expo >0 else expo-1
    manti=np.round(float(a)/10**expo, precision)
    if expo <-1 or expo > 1:
        return '{}E{}'.format(manti ,expo)
    else:
        return '{}'.format(np.round(a,precision))
    
```

### 3.7

Let's begin by defining the two composite integral approximation functions.

```python
def composite_trapezoidal(f, a, b, n):
    h = (b - a) / n
    result = h * (sum([f(a + j*h) for j in range(n+1)]) - 0.5 * (f(a) + f(b)))
    return result
```

```python
def composite_simpson(f, a, b, n):
    h = (b - a) / (2 * n)
    result = 0
    for i in range(n):
        result += h/3 * (f(a + (2*i)*h) + 4 * f(a + (2*i+1)*h) + f(a + (2*i+2)*h))
    return result
```

#### a)

Now let's use the composite rules to evaluate the function $f(x) = \tan(x)$ over $[0,1]$ with 10 subintervals. The results are displayed in the dataframe below.

```python
f = lambda x: np.tan(x)

true = quad(f, 0, 1)[0]
trap = composite_trapezoidal(f, 0, 1, 10)
simp = composite_simpson(f, 0, 1, 10)

df = pd.DataFrame({
    'Type': ['Comp Trapezoid', 'Comp Simpson', 'True'],
    'Value': [trap, simp, true],
})

df['Error'] = df['Value'] - true
df.replace(0, np.nan, inplace=True)

display(df)
```

#### b)

```python
trap_errors = []
simp_errors = []
n_vals = [2**k for k in range(11)]

for n in n_vals:
    true = quad(f, 0, 1)[0]
    trap = composite_trapezoidal(f, 0, 1, n)
    simp = composite_simpson(f, 0, 1, n)
    trap_errors.append(abs(trap-true))
    simp_errors.append(abs(simp-true))

h_vals = [1/n for n in n_vals]

plt.plot(np.log(h_vals), np.log(trap_errors), label='composite trapezoid')
plt.plot(np.log(h_vals), np.log(simp_errors), label='composite simpson')
plt.plot(np.log(h_vals), [np.log(h**2) for h in h_vals], '--', label='h^2')
plt.plot(np.log(h_vals), [np.log(h**4) for h in h_vals], '--', label='h^4')
plt.title('Quadrature error for tan(x)')
plt.xlabel('log(h)')
plt.ylabel('log(y)')
plt.legend()
plt.show()
```

We can see above that the composite trapezoidal rule has quadrature error of order 2, parallel to $h^2$, and the composite simpson rule has quadrature error of order 4, parallel to $h^4$.

#### c)

```python
df_errors = pd.DataFrame({
    '$\tilde{h}$': h_vals,
    'error $I_{1,n}$': trap_errors,
    'eval $I_{1,n}$': [n+1 for n in n_vals],
    'error $I_{2,n}$': simp_errors,
    'eval $I_{2,n}$': [2*n+1 for n in n_vals],
})

display(df_errors)
```

We can see that composite simpson is more efficient, requiring 33 function evaluations to reduce the error down to below $10^{-6}$, whereas composite trapezoidal requires 513 evaluations.

```python
def forward_euler(alpha, n, t_final):
    # method for f = alpha*y(t)
    h = t_final / n
    u = np.zeros(n+1)
    t = np.linspace(0, t_final, n+1)
    u[0] = 1
    for i in range(n):
        u[i+1] = u[i] + h * alpha * u[i]
    return t, u

def exact(alpha, t):
    # solution of y'(t) = alpha*y(t)
    return np.exp(alpha * t)

alphas = [-0.8, -0.1, 0.1, 0.3]

for alpha in alphas:
    t, u_euler = forward_euler(alpha, 50, 5)
    u_exact = exact(alpha, t)
    plt.plot(t, u_euler, label=f'Euler, alpha={alpha}')
    plt.plot(t, u_exact, '--', label=f'Exact, alpha={alpha}')

plt.title('Forward Euler Method vs Exact Solution')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

#### b)

```python
def heun(alpha, n, t_final):
    # method for f = alpha*y(t)
    h = t_final / n
    u = np.zeros(n+1)
    t = np.linspace(0, t_final, n+1)
    u[0] = 1
    fe = u.copy()
    for i in range(n):
        fe[i+1] = fe[i] + h * alpha * fe[i]
        u[i+1] = u[i] + h/2 * (alpha * u[i] + alpha * fe[i+1])
    return t, u

for alpha in alphas:
    t, u_heun = heun(alpha, 50, 5)
    u_exact = exact(alpha, t)
    plt.plot(t, u_heun, label=f'Heun, alpha={alpha}')
    plt.plot(t, u_exact, '--', label=f'Exact, alpha={alpha}')

plt.title('Heun Method vs Exact Solution')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
